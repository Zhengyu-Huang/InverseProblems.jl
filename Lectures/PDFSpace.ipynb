# Probability Density Function Space 

## Spaces

* Density function space on $\theta \in \Omega$: 

$$\mathcal{P}(\Omega) = \Bigl\{\rho \in \mathcal{C}^{\infty}(\Omega) \cap L_1(\Omega), \rho > 0 \textrm{ a.e.}, \int \rho d\theta = 1 \Bigr\} $$

* Tangent space of $\mathcal{P}$ at $\rho$:
$$
\begin{align*}
T_{\rho}\mathcal{P} = \bigl\{\frac{d}{dt}\rho(t)\Big|_{t=0} : \rho(t) \textrm{ is a curve in } \mathcal{P}, \rho(0) = \rho \bigr\} = \bigl\{\sigma \in \mathcal{C}^{\infty}(\Omega), \int \sigma d\theta = 0 \bigr\}
\end{align*}
$$
It is worth mentioning that $t$ can be any parameterization, and $T_{\rho}\mathcal{P}$ is a vector space.

* Cotangent space of $\mathcal{P}$ at $\rho$:
$$
\begin{align*}
T_{\rho}^{*}\mathcal{P} = \bigl(T_{\rho}\mathcal{P}\bigr)^{*}:  T_{\rho}\mathcal{P} \rightarrow \mathbb{R}
\end{align*}
$$
Namely, elements of the cotangent space $T_{\rho}^{*}\mathcal{P}$ are linear functionals on $T_{\rho}\mathcal{P}$.




## Functions

Let define any function on the density function space $\mathcal{P}(\Omega)$
$$
\mathcal{E}(\rho) :\quad \rho \rightarrow \mathbb{R} T_{\rho}^{*}\mathcal{P}$
$$

The functional derivative of $\mathcal{E}(\rho)$ at $\rho$ is

$$
\frac{\delta \mathcal{E}}{\delta \rho}\Big|_{\rho} \in T_{\rho}^{*}\mathcal{P}
$$

Its action on $\sigma \in T_{\rho}\mathcal{P}$ is 

$$
\frac{\delta \mathcal{E}}{\delta \rho}\Big|_{\rho} \sigma  = \lim_{\epsilon \rightarrow 0}\frac{\mathcal{E}(\rho + \epsilon \sigma) - \mathcal{E}(\rho)}{\epsilon}
$$

$T_{\rho}^{*}\mathcal{P}$ is a linear space on $T_{\rho}\mathcal{P}$ 

$$
\frac{\delta (\mathcal{E}_1 + \mathcal{E}_2)}{\delta \rho} \delta = 
\frac{\delta \mathcal{E}_1 }{\delta \rho} \delta + 
\frac{\delta \mathcal{E}_2}{\delta \rho}\delta
$$


## Metric tensor

The metric tensor at the point $\rho$ is essentially a **bilinear** function

$$
g_{\rho}(\sigma_1, \sigma_2) : T_{\rho}\mathcal{P} \times T_{\rho}\mathcal{P} \rightarrow \mathbb{R}
$$

It needs to be **symmetric**, 

$$
g_{\rho}(\sigma_1, \sigma_2)  = g_{\rho}(\sigma_2, \sigma_1) 
$$

and **nondegenerate**, namely, for every $\sigma_1 \neq 0$, there exists a $\sigma_2$, such that $g_{\rho}(\sigma_1, \sigma_2) \neq 0$. When the metric tensor is **positive definite**, 

$$
g_{\rho}(\sigma, \sigma) \geq 0, \textrm{ with equality if and only if } \sigma = 0, 
$$

it is called a **Riemannian metric**. In this lecture, we focus only on **Riemannian metric**, which makes a Hilbert space on $T_{\rho}\mathcal{P}$. 

### Another perspective
Consider the linear functional on $T_{\rho}\mathcal{P}$, 

$$
g_{\rho}(\sigma_1, \,\cdot\,) : \quad T_{\rho}\mathcal{P} \rightarrow \mathbb{R}
$$

There exists $\Phi \in T_{\rho}^{*}\mathcal{P}$, such that 

$$
\Phi \sigma_2 = g_{\rho}(\sigma_1, \sigma_2).
$$

Therefore, the metric tensor can be defined as 

$$
G(\rho) : T_{\rho}\mathcal{P} \rightarrow T_{\rho}^{*}\mathcal{P}
$$

such that 

$$
g_{\rho}(\sigma_1, \sigma_2)  = G(\rho)(\sigma_1) \sigma_2
$$

The symmetry and positiveness lead to that 

$$
\begin{align*}
&G(\rho)(\sigma_1) \sigma_2 = g_{\rho}(\sigma_1, \sigma_2) =  g_{\rho}(\sigma_2, \sigma_1) = G(\rho)(\sigma_2) \sigma_1\\
&G(\rho)(\sigma) \sigma \geq 0, \textrm{ with equality if and only if } \sigma = 0. 
\end{align*}
$$

Moreover, for Riemannian metric, 
* $T_{\rho}\mathcal{P}$ is a Hilbert space with the inner product $g_{\rho}(\,\cdot,\,\cdot\,)$. 

* $T_{\rho}\mathcal{P}$ and $T_{\rho}^{*}\mathcal{P}$ are isomorphic. 

* Based on Riesz representation theorem,

$$
\forall \Phi \in T_{\rho}^{*}\mathcal{P},\quad \exists \sigma_1 \in T_{\rho}\mathcal{P} \quad\textrm{s.t.} \quad \Phi(\,\cdot\,) = g_{\rho}(\sigma_1, \,\cdot\,).
$$

Therefore $G(\rho)$ is bijective, and we have 
$$
G(\rho)^{-1} : T_{\rho}^{*}\mathcal{P} \rightarrow T_{\rho}\mathcal{P}
$$

## Distance

After defining metric tensor, we can define arc length of $\Gamma: [0,\,1]\rightarrow \rho(t)$, as follows, 

$$
\begin{align*}
\int_{t=0}^{t=1} \sqrt{g_{\rho}\bigl(\frac{\partial \rho(t)}{\partial t}, \frac{\partial \rho(t)}{\partial t}\bigr)} dt
\end{align*}
$$

The distance between 2 densities $\rho^A$ and $\rho^B$ is the smallest arc length between them. However for simplicity, the 2-distance is usually used

$$
\begin{align*}
&{\rm dist}(\rho^A, \rho^B) = \min_{\sigma \in T_{\rho}\mathcal{P}} \int_{t=0}^{t=1} g_{\rho}(\sigma, \sigma) dt\\
&s.t. \qquad \frac{\partial \rho(t)}{\partial t} =  \sigma \qquad \rho(0) = \rho^A \quad \rho(1) = \rho^B
\end{align*}
$$

where we use $g_{\rho}$ instead of $\sqrt{g_{\rho}}$. It is worth mentioning that ${\rm dist}(\rho^A, \rho^B)$ is a distance, since

* ${\rm dist}(\rho^A, \rho^B) = {\rm dist}(\rho^B, \rho^A)$
* ${\rm dist}(\rho^A, \rho^B) = 0 \iff \rho^A = \rho^B$
* ${\rm dist}(\rho^A, \rho^C) \leq  {\rm dist}(\rho^A, \rho^B) + {\rm dist}(\rho^B, \rho^C)$

The distance can also be rewritten as 

$$
\begin{align*}
&{\rm dist}(\rho^A, \rho^B) = \min_{\Phi \in T_{\rho}^{*}\mathcal{P}} \int_{t=0}^{t=1} g_{\rho}(G(\rho)^{-1}\Phi, G(\rho)^{-1}\Phi) dt\\
&s.t. \qquad \frac{\partial \rho(t)}{\partial t} =  G(\rho)^{-1}\Phi \qquad \rho(0) = \rho^A \quad \rho(1) = \rho^B
\end{align*}
$$

## What is $T_{\rho}^{*}\mathcal{P}$?  A formal representaion 

Consider the Hilbert space $T_{\rho}\mathcal{P}$ with the $L_2$ inner product:
$$
\langle \sigma_1, \sigma_2 \rangle = \int \sigma_1  \sigma_2 d \theta
$$

Assume that $\Phi \in T_{\rho}^{*}\mathcal{P}$ is bounded 
$$
\Phi \sigma = g_{\rho}(\sigma, \sigma) \leq C  \langle \sigma, \sigma \rangle
$$

Then following Riesz representationa theorem, we have $\Phi^{\sigma}$, the representaion of $\Phi$ in $T_{\rho}\mathcal{P}$, and its action on $\sigma \in T_{\rho}\mathcal{P}$ 

$$
\begin{align*}
\Phi \sigma = \int \Phi^{\sigma} \sigma d\theta
\end{align*}
$$

In this lecture, we might abuse the notation by using $\Phi = \Phi^{\sigma}$.

# Application

## Gradient descent in probability density function space

Let define the Kullback–Leibler divergence function on the density space $\mathcal{P}(\Omega)$
$$
\begin{align*}
\mathcal{E}(\rho) = KL\Bigl[\rho \Big\Vert  \rho_{\rm post}(\theta | y)\Bigr] = \int\rho \log \rho_{\rm post}(\theta | y)d\theta - \int\rho\log \rho d\theta
\end{align*}
$$

We can define the functional derivative 

$$
\begin{align*}
\frac{\delta \mathcal{E}}{\delta \rho} \sigma 
&= \int \sigma \log \rho_{\rm post}(\theta | y)d\theta - \int \sigma \log \rho d\theta - \int\sigma \rho \delta \log\rho  d\theta \\
&= \int \sigma \Bigl(\log \rho_{\rm post}(\theta | y) - \log \rho - 1 \Bigr) d\theta
\end{align*}
$$

Following Riesz representationa theorem, we have the representaion of $\frac{\delta \mathcal{E}}{\delta \rho}$ in $T_{\rho}\mathcal{P}$:

$$
\begin{align*}
\frac{\delta \mathcal{E}}{\delta \rho} 
= \log \rho_{\rm post}(\theta | y) - \log \rho - 1 - C
\end{align*}
$$
where $C$ is the normalization constant.

For any metric tensor $G(\rho)$, then the steepest gradient flow to minimize $\mathcal{E}$ becomes

$$\frac{\partial \rho_t}{\partial t} = - G(\rho)^{-1} \frac{\delta \mathcal{E}}{\delta \rho}$$

since

$$\frac{\partial \mathcal{E}}{\partial t} = \frac{\delta \mathcal{E}}{\delta \rho_t} \frac{\partial \rho_t}{\partial t} = - \frac{\delta \mathcal{E}}{\delta \rho_t} G(\rho_t)^{-1} \frac{\delta \mathcal{E}}{\delta \rho_t} = g_{\rho}\bigl(G(\rho_t)^{-1} \frac{\delta \mathcal{E}}{\delta \rho_t}, G(\rho_t)^{-1} \frac{\delta \mathcal{E}}{\delta \rho_t}\bigr) $$




## Example: 2-Wasserstein metric 

$$
G(\rho)^{-1} \Phi = - \nabla\cdot(\rho\nabla\Phi), \qquad  \Phi \in T_{\rho}^{*}\mathcal{P}
$$

The distance between $\rho^A$ and $\rho^B$ becomes

$$
\begin{align*}
{\rm dist}(\rho^A, \rho^B) 
&= \min_{\Phi \in T_{\rho}^{*}\mathcal{P}} \int_{t=0}^{t=1} g_{\rho_t}(G(\rho_t)^{-1} \Phi, G(\rho_t)^{-1} \Phi) dt  \\
&= \min_{\Phi \in T_{\rho}^{*}\mathcal{P}} \int_{t=0}^{t=1} \int -\Phi \nabla\cdot(\rho_t\nabla\Phi) d\theta dt  \\
&= \min_{\Phi \in T_{\rho}^{*}\mathcal{P}} \int_{t=0}^{t=1} \int \rho_t \nabla \Phi \cdot \nabla \Phi d\theta dt\\
&s.t. \qquad \frac{\partial \rho_t}{\partial t} + \nabla\cdot(\rho_t\nabla\Phi) = 0  \qquad \rho_0 = \rho^A \quad \rho_1 = \rho^B
\end{align*}
$$

This corresponds to the Benamou and Brenier formulation [1] of Wasserstein-2 distance. The original definition of Wasserstein-2 distance is 

$$
\begin{align*}
&\min_\rho \int \frac{1}{2}\lVert \theta_1 - \theta_2 \rVert^2 \rho(\theta_1 , \theta_2)d\theta_1 d\theta_2 \quad 
\textrm{such that} \quad  \rho^A = \int\rho(\theta_1, \theta_2) d\theta_2 \quad 
\rho^B = \int\rho(\theta_1, \theta_2) d\theta_1 \\
\end{align*}
$$

For the Gradient descent of KL divergence, the density evolution equation is

$$
\begin{align*}
\frac{\partial \rho_t(\theta)}{\partial t} 
&= -\nabla_{\theta} \cdot \Bigl[\rho_t \cdot \nabla_{\theta}\bigl(\log \rho_{\rm post}(\theta | y) - \log \rho_t - 1\bigr) \Bigr]\\
&= -\nabla_{\theta} \cdot \Bigl[\rho_t \cdot \nabla_{\theta}\log \rho(\theta,y)\Bigr] + \nabla_{\theta}\Bigl[ \nabla_{\theta} \rho_t \Bigr]
\end{align*}
$$

which corresponds to Langevin dynamics.

## Example: Fisher-Rao metric

$$
G(\rho)^{-1} \Phi = \rho\Phi - \rho\mathbb{E}_\rho[\Phi], \qquad  \Phi \in T_{\rho}^{*}\mathcal{P}
$$

The distance between $\rho^A$ and $\rho^B$ becomes

$$
\begin{align*}
{\rm dist}(\rho^A, \rho^B) &= \min_{\Phi \in T_{\rho}^{*}\mathcal{P}}  \int_{t=0}^{t=1} g_{\rho_t}(G(\rho_t)^{-1} \Phi, G(\rho_t)^{-1} \Phi) dt   \\
&= \min_{\Phi \in T_{\rho}^{*}\mathcal{P}} \int_{t=0}^{t=1} \int \rho_t \Phi (\Phi - \mathbb{E}_{\rho_t}[\Phi]) d\theta dt\\
&s.t. \qquad \frac{\partial \rho_t}{\partial t} - \rho_t (\Phi - \mathbb{E}_{\rho_t}[\Phi]) = 0 \qquad \rho_0 = \rho^A \quad \rho_1 = \rho^B
\end{align*}
$$

For the Gradient descent of KL divergence, the density evolution equation becomes

$$
\begin{align*}
\frac{\partial \rho_t\bigl(\theta)}{\partial t} 
&= \rho_t (\frac{\delta \mathcal{E}}{\delta \rho_t} - \mathbb{E}_{\rho_t}[\frac{\delta \mathcal{E}}{\delta \rho_t}]\bigr) \\
&= \rho_t \bigl(\log \rho(\theta, y) - \log \rho_t(\theta) - \mathbb{E}_{\rho_t}[\log \rho(\theta, y) - \log \rho_t(\theta)]\bigr)
\end{align*}
$$

We have 

$$
\begin{align*}
\frac{\partial \log \rho_t(\theta)}{\partial t} 
&= \log \rho(\theta, y) - \log \rho_t(\theta) - \mathbb{E}_{\rho_t}[\log \rho(\theta, y) - \log \rho_t(\theta)]\\
\frac{\partial e^t \log \rho_t(\theta)}{\partial t} 
&= e^t\log \rho(\theta, y) - e^t\mathbb{E}_{\rho_t}[\log \rho_t(\theta, y) - \log \rho_t(\theta)]\\
\log \rho_t(\theta)
&= (1-e^{-t})\log \rho(\theta, y) + e^{-t}\log\rho_0 - C_t \qquad C_t = \int_0^{t} e^{\tau-t}\mathbb{E}_{\rho_\tau}[\log \rho(\theta, y) - \log \rho_\tau(\theta)] d\tau 
\end{align*}
$$

Therefore, we have the "fixed point iteration"

$$
\begin{align*}
\rho_t \propto \rho_0^{e^{-t}} \rho(\theta, y)^{1 - e^{-t}}
\end{align*}
$$
